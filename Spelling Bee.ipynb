```python
word_list = set()
with open('sowpods.txt', 'r') as infile:
    for line in infile:
        if len(line) >= 5:
            word_list.add(line.rstrip())
    
```

```python
def count(word_list, puzzles):
    result = [0] * len(puzzles)
    puzzles = [Puzzle(p) for p in puzzles]
    words = [set([c for c in word]) for word in word_list]
    for i in range(len(puzzles)):
        puzzle = puzzles[i]
        for word in words:
            if puzzle.valid(word):
                result[i] += 1
    return result
        
class Puzzle:
    def __init__(self, puzzle):
        self.key = puzzle[0]
        self.all_chars = set([c for c in puzzle])
        
    def valid(self, word):
        has_key = False
        for c in word:
            if not c in self.all_chars:
                return False
            elif c == self.key:
                has_key = True
        return has_key
```

```python
count(word_list, puzzles)
```
